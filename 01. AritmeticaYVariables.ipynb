# Variables en Python

Las variables son contenedores que almacenan datos. En Python, no es necesario declarar el tipo de variable, ya que se asigna automáticamente según el valor que se le asigne.

```python
# Create a variable called test_var and give it a value of 4+5
test_var = 4 + 5

# Print the value of test_var
print(test_var)
```

```python
# Set the value of a new variable to 3
my_var = 3

# Print the value assigned to my_var
print(my_var)

# Change the value of the variable to 100
my_var = 100

# Print the new value assigned to my_var
print(my_var)
```

```python
print(my_var)
print(test_var)
```

```python
# Increase the value by 3
my_var = my_var + 3

# Print the value assigned to my_var
print(my_var)
```

```python
# Create variables
num_years = 4
days_per_year = 365 
hours_per_day = 24
mins_per_hour = 60
secs_per_min = 60

# Calculate number of seconds in four years
total_secs = secs_per_min * mins_per_hour * hours_per_day * days_per_year * num_years
print(total_secs)
```

```python
# Update to include leap years
days_per_year = 365.25

# Calculate number of seconds in four years
total_secs = secs_per_min * mins_per_hour * hours_per_day * days_per_year * num_years
print(total_secs)
```
