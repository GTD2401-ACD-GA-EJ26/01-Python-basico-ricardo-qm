# Introducción a Listas

En este JNB se revisará el tema de Introducción a Listas

```python
flowers = "pink primrose,hard-leaved pocket orchid,canterbury bells,sweet pea,english marigold,tiger lily,moon orchid,bird of paradise,monkshood,globe thistle"

print(type(flowers))
print(flowers)
```

```python
flowers_list = ["pink primrose", "hard-leaved pocket orchid", "canterbury bells", "sweet pea", "english marigold", "tiger lily", "moon orchid", "bird of paradise", "monkshood", "globe thistle"]

print(type(flowers_list))
print(flowers_list)
```

```python
# The list has ten entries
print(len(flowers_list))
```

```python
print("First entry:", flowers_list[0])
print("Second entry:", flowers_list[1])

# The list has length ten, so we refer to final entry with 9
print("Last entry:", flowers_list[9])
```

```python
# The list has length ten, so we refer to final entry with -1
print("Last entry:", flowers_list[-1])
```

```python
print("First three entries:", flowers_list[:3])
print("Final two entries:", flowers_list[-2:])
```

```python
print("Entries from 3 to 6:", flowers_list[3:7])
```

```python
# Imprime la lista en sentido inverso
print("Entries in reverse order:", flowers_list[::-1])
```

```python
# Imprime los pares de la lista desde el tres hasta el ocho de dos en dos
print("Entries from 3 to 8 with step 2:", flowers_list[3:9:2])
```

```python
# Haz lo mismo pero en sentido contrario
print("Entries from 8 to 3 with step -2:", flowers_list[8:2:-2])
```

```python
# Haz el mismo ejercicio anterior pero con indices negativos

print("Entries from -7 to -2 with step 2:", flowers_list[-7:-1:2])
```
